```python
import operator
import math

TotalT = 0
CuentaF = 0
CuentaG = 0
CuentaH = 0
CuentaI = 0
CuentaJ = 0
CuentaT = 0
TotalF = 0
TotalG = 0
TotalH = 0
TotalI = 0
TotalJ = 0
l = 1

rut_list = []
num_cal = [3, 2, 7, 6, 5, 4, 3, 2]
RPE = []

def dv(n):
    x = sum(n)
    e = math.floor(x / 11)
    f = x - (11 * e)
    h = 11 - f

    if h == 11:
        h = "0"
    elif h == 10:
        h = "k"
    return str(h)

rut_list = []

import numpy as np

# Crear un mapa vacío de departamentos
mapa_departamentos = np.zeros((10, 4), dtype=str)

# Función para mostrar el mapa de departamentos
def mostrar_mapa():
    print("Mapa de Departamentos")
    print("       A   B   C   D")
    print("---------------------")
    for fila in range(10):
        fila_str = f"Piso {10 - fila:2d} |"
        for columna in range(4):
            fila_str += f" {mapa_departamentos[fila, columna]} |"
        print(fila_str)

# Función para marcar un departamento como comprado
def marcar_departamento(fila, columna):
    mapa_departamentos[fila, columna] = "X"

# Función para mostrar el menú
def mostrar_menu():
    print("Menú:")
    print("1. Comprar departamento")
    print("2. Mostrar precios de compra")
    print("3. Mostrar listado de departamentos disponibles")
    print("4. Mostrar listado de compradores")
    print("5. Mostrar ganancias totales")
    print("6. Salir")

# Variable para controlar el flujo del programa
comprar = True
Clientes = {}

while comprar:
    mostrar_menu()
    opcion = int(input("Ingrese la opción deseada: "))

    if opcion == 1:
        print("COMPRANDO DEPARTAMENTO")
        mostrar_mapa()
        while True:
            fila = int(input("Ingrese el piso del departamento a comprar (1-10): "))
            if 1 <= fila <= 10:
                break
            else:
                print("Error: opción inválida.")

        while True:
            columna = ord(input("Ingrese la columna del departamento a comprar (A-D): ")) - 65

            if 0 <= columna <= 3:
                break
            else:
                print("Error: opción inválida.")

        if mapa_departamentos[10 - fila, columna] == "X":
            print("No válido para compra. Departamento ya comprado.")
            continue

        rut = input("Ingrese su rut para validar si es apto y iniciar la compra, sin dígito verificador (para cancelar solo ingrese 0): ")
        if rut == '0':
            continue
        else:
            try:
                rut_list = [int(x) for x in rut]
                calculo = list(map(operator.mul, rut_list, num_cal))
                Dvc = int(input("Por favor, ingrese el dígito verificador de su rut: "))
                if str(Dvc) == dv(calculo):
                    print("Es apto el RUT")
                    NP = input("Ingrese el nombre del cliente con apellido: ")
                    NP_lower = NP.lower()
                    Clientes[NP_lower] = [rut, Dvc, fila, columna]
                    marcar_departamento(10 - fila, columna)
                    mostrar_mapa()
                    print("MAPA ACTUALIZADO")

                    if columna == 0:
                        TotalF += 3800
                        CuentaF += 1
                    elif columna == 1:
                        TotalG += 3000
                        CuentaG += 1
                    elif columna == 2:
                        TotalH += 2800
                        CuentaH += 1
                    elif columna == 3:
                        TotalI += 3500
                        CuentaI += 1
                    CuentaT += 1
                    TotalT = TotalF + TotalG + TotalH + TotalI

                else:
                    print("RUT inválido, ingrese otro.")
            except ValueError:
                print("Por favor, ingrese un RUT válido.")

    elif opcion == 2:
        print("PRECIOS DE COMPRA:")
        print("A\t3800 UF")
        print("B\t3000 UF")
        print("C\t2800 UF")
        print("D\t3500 UF")

    elif opcion == 3:
        print("MOSTRANDO LISTADO DE DEPARTAMENTOS DISPONIBLES")
        mostrar_mapa()

    elif opcion == 4:
        print("MOSTRANDO LISTADO DE COMPRADORES")
        print("Cliente\t\tRUT\tDV\tPiso\tColumna")
        for cliente, datos in Clientes.items():
            print(f"{cliente}\t{datos[0]}\t{datos[1]}\t{datos[2]}\t{chr(datos[3] + 65)}")

    elif opcion == 5:
        print("MOSTRANDO GANANCIAS TOTALES")
        print(f"TIPO\tPRECIO\tCANTIDAD\tTOTAL")
        print(f"A\t3800 UF\t{CuentaF}\t{TotalF}")
        print(f"B\t3000 UF\t{CuentaG}\t{TotalG}")
        print(f"C\t2800 UF\t{CuentaH}\t{TotalH}")
        print(f"D\t3500 UF\t{CuentaI}\t{TotalI}")
        print("----------------------")
        print(f"Total departamentos vendidos:\t{CuentaT}")
        print(f"Total ganancias:\t\t{TotalT}")

    elif opcion == 6:
        print("Saliendo")
        comprar = False

print("Nombre Programador: Daniel Plasencia Fecha 11/07/2023")
```
